1. Set the spark config for SAS token
1. List files from demo container
1. Read data from circuits.csv file


```python
spark.conf.set("fs.azure.account.auth.type.formula1dl1216.dfs.core.windows.net", "SAS")
spark.conf.set("fs.azure.sas.token.provider.type.formula1dl1216.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.formula1dl1216.dfs.core.windows.net", "sp=rl&st=2025-07-23T15:48:51Z&se=2025-07-24T00:03:51Z&spr=https&sv=2024-11-04&sr=c&sig=%2BG0XOWGeq2zPfdImsiuDhRH5jJuNZHLr3vpedefmKD0%3D")
```
